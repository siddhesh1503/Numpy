```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("=== Zomato sales analysis ===")
print("\n Sales Data Shape", sales_data.shape)
print("\n Sample data for 1st restaurant:\n", sales_data[0 : 3]) 
```

```python
# total sales per year 
print(np.sum(sales_data, axis=0)) #axis=0 means columns wise and axis=1 is row wise
yearly_total = np.sum(sales_data[:, 1:], axis=0)
print(yearly_total)

# Minimum sales per restaruant
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)

# Maximum sales per years
max_sales = np.max(sales_data[:, 1:], axis=0)
print(max_sales)

# Average sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print(avg_sales)
```

```python

cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)
# You can plot cumulative sales for each restaurant
for i in range(cumsum.shape[0]):
    plt.plot(cumsum[i], label=f'Restaurant {i+1}')
plt.legend()
plt.title("Cumulative sales per restaurant")
plt.xlabel("Years")
plt.ylabel("Cumulative Sales")
plt.grid(True)
plt.show()
plt.figure(figsize=(8, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales accross all restaurant ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector Addition: ", vector1 + vector2)
print("\nVector Multiplication: ", vector1 * vector2)
print("\nDot Product: ", np.dot(vector1, vector2))

cos_theta = np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))
angle = np.arccos(cos_theta)
print("\nAngle (radians):", angle)
```
